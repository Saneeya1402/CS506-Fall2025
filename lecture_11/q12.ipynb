```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python

X = [[0,0],[0,4],[4,0],[0,-4],[-4,0],[0,1],[0,-7],[0,-1],[0,-7],[0,-2],[0,-6],[0,-3],[0,-5],[1,0],[7,0],[2,0],[6,0],[3,0],[5,0],[-1,0],[-7,0],[-2,0],[-6,0],[-3,0],[-5,0],[0,3.95],[0,4.05]]
X = np.array(X)
eps = 1
min_samples = 2
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
labels = dbscan.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = ['b', 'g']

for label in unique_labels:
    color = 'k' if label == -1 else colors[label % len(colors)]
    plt.scatter(X[labels == label, 0], X[labels == label, 1], c=color, label=f"Cluster {label}")

plt.title("DBSCAN Clustering")
plt.legend()
plt.show()
```

```python
sil_scores = silhouette_samples(X, labels)
print(sil_scores)
print('Silhouette Score of', X[0], 'is: ' + str(sil_scores[0]))
```
